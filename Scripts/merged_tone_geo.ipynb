```python
import pandas as pd
```

```python
text = pd.read_csv('../Raw_Data/Hydrated/cleaned_geo_text.csv', index_col=0, dtype={"id": object})
```

```python
p_text = pd.read_csv("../Raw_Data/Hydrated/cleaned_geo_with_processed_text.csv", index_col=0, dtype={"id": object})
```

```python
tones = pd.read_csv("../Tone_Analyzer/Data/combined_data.csv")
```

```python
text.head()
```

```python
p_text = p_text.rename(columns={"id": "id",
                                "text": "processed_text",
                                "coordinates": "coordinates"
                                                })
p_text
```

```python
merge_df = pd.merge(text, p_text, on="id")
merge_df = merge_df.drop(columns=['coordinates'])
merge_df
```

```python
tones = tones.rename(columns={"date": "date",
                                "text": "processed_text",
                                "tone": "tone",
                                "score": "score"
                                                })
tones
```

```python
geo_tones = pd.merge(tones, merge_df, on="processed_text")
geo_tones = geo_tones.drop(columns=['text'])
geo_tones
```

```python
geo_tones.to_csv("../Raw_Data/Hydrated/geo_tones.csv")
```
